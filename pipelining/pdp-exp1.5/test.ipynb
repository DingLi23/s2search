```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
from math import sqrt

x_len = 2
y_len = 2

inter = 0

def y(p):
    return 10 * p['x1'] + 4 * p['x2'] + (inter * (pow(p['x1'], 2) + sqrt(p['x2'])))

x1_v = [5, 3,2,10,12,5,4,1,4,1,2,6,6,7,6]
x2_v = [2, 4,4,12,3,4,6,2,3,5,1,2,3,5,3,1]

paper_d = []

for i in range(len(x1_v)):
    paper_d.append({
        'x1': x1_v[i],
        'x2': x2_v[i]
    })

score_map = []

for i in range(len(paper_d)):
    for p in paper_d:
        v_data = []
        for j in range(len(paper_d)):
            new_data = {}
            new_data['x1'] = paper_d[j]['x1']
            new_data['x2'] = paper_d[i]['x2']
            v_data.append(new_data)
            
#     for p in v_data:
#         print(p)
        
    score_map.append(list(map(lambda x: y(x), v_data)))
    
# print(score_map)


norm = mcolors.Normalize(vmin=0., vmax=100.)

# pc_kwargs = {'rasterized': True, 'cmap': 'viridis', 'norm': norm}

fig, ax = plt.subplots(figsize=(10, 10), constrained_layout=True)

im = ax.pcolormesh(score_map)

fig.colorbar(im, ax=ax, shrink=0.6)
```
